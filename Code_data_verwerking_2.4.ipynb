```python
#importeren van de benodige lybarieries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# naam van het Excel-bestand
excel_bestand = 'blokdak.xlsx'

# Lees het Excel-bestand in met pandas
df = pd.read_excel(excel_bestand)

# Kies de gewenste rij
rij_index = 0
rij = df.iloc[rij_index]
# Toon de inhoud van de rij
print(rij)
```

```python
#zet de coordinaten van de meetpunten er in(x is lengte en y is breete )
x = [5.2,16.675,28.15,37.225,46.3,5.2,16.675,28.15,37.225,46.3,51.1,5.2,16.675,28.15,37.225,46.3]
y = [4,4,4,4,4,13.35, 13.35,13.35,13.35,13.35, 13.35, 20.9, 20.9,20.9,20.9,20.9]
df3 = pd.DataFrame(np.array([x,y]).T, columns=['x','y'])
df3
```

```python
#frequentie stappen(belengrijk voor de titel)
start_frequentie = 100
eind_frequentie = 20000
stappen = 20
frequenties =np.logspace(np.log10(start_frequentie), np.log10(eind_frequentie), stappen)
print(frequenties)
woord='Hz'
```

```python
#maakt een mapje aan genaamt images
!mkdir images
```

```python
#maakt de lengte voor het for loopje aan
#in dit geval is het 20 omdat er 20 frequenties zijn
B=np.zeros(20) 
for i in range(len(B)):
    rij_index = i
    rij = df.iloc[rij_index]
    # Maak een scatterplot met kleurbar
    scatter = plt.scatter(df3.x, df3.y, c=rij, cmap='rainbow', marker='.', s=2000)
    colorbar = plt.colorbar(scatter, label="dB")
    colorbar.set_ticks(np.linspace(rij.min(), rij.max(), num=16))
    plt.clim(40, 110)
    #labels voor x-as en y-as
    plt.xlabel("x[cm]")
    plt.ylabel("y[cm]")
    #liemieten kamer 
    plt.xlim(0, 56.3)
    plt.ylim(0, 26.7)
    # titel voor het makelijk identificeren van de gis's
    number= round(frequenties[i])
    title = "{}: {} {}".format(excel_bestand,number,woord)
    plt.title(title)
    #opslaan van het scatter plot als png in images map
    plt.savefig(f'images/{i:003}', dpi='figure', format=None, metadata=None,
                bbox_inches=None, pad_inches=0.1,
                facecolor='white', edgecolor='auto',
                backend=None)
    
    plt.show()
```
